### beautifulsoup 라이브러리
1. html tag로 이루어진 문자열 데이터를 파싱하여  데이터를 쉽게 추출하기 위한 라이브러리이다.
2. 웹의 구조를 어느정도 인지하고 있다면 쉽게 사용할 수 있다.
3. parser를 활용하면 파이썬에서 tag에 접근이 쉽도록 객체 형태로 제공한다.


```python
#parser를 이용하여 데이터를 변환
soup = bs(html_data, 'html.parser')
```

```python
# 태그의 이름을 사용하여 검색
    # soup.태그명 : 해당하는 태그들 중 첫번째의 정보를 출력
    # soup.태그명.string : 검색한 태그의 문자열을 출력해줌
    # soup.태그명[{속성}] : 검색한 태그의 속성의 값 출력
print(soup.title)
print(soup.title.string)
```

```python
print(soup.a)
print(soup.a['href'])
```

1. find()
    - HTML 문자열 안에서 해당 태그의 첫번째 정보를 출력
    - find(속성 = 값): 태그들 중 해당 속성의 값이 일치하는 태그의 정보를 출력.
2. find_all()
    - HTML 문자열 안에서 해당 태그의 모든 정보를 리스트 형태로 출력.
    - limit 매개변수: 출력되는 리스트의 길이를 지정


```python
# 태그의 정보 확인
soup.find('a')
```

```python
soup.find_all('a')
```

```python
# 네이버에 있는 a태그의 contents의 값들을 리스트로 생성하려면?

# case1 (for)
# 비어있는 리스트를 생성
a_contents = []

# 반복문을 이용하여 a태그의 모든 정보 리스트를 반복
for i in soup.find_all('a'):
    # i?  → 태그 하나의 정보
    a_content = i.get_text()
    a_contents.append(a_content)


print(a_contents)
```

```python
# case2 (map)
# map(함수명 또는 lambda 함수, 자료형데이터)

list(
    map(

        lambda x : x.get_text(), 
        soup.find_all('a')
    ))
```

```python
response = requests.get('https://search.shopping.naver.com/search/all?where=all&frm=NVSCTAB&query=%ED%82%A4%EB%B3%B4%EB%93%9C')
```

```python
html_data = response.text
```

```python
soup = bs(html_data, 'html.parser')
```

```python
soup
```

```python
#div 태그 중 class의 값이 특정한 class인 태그를 검색
div_data = soup.find(
    'div',
    attrs={
        'class' : "style_content__xWg5l"
    }
)
```

```python
div_data
```

```python
ad_list = div_data.find_all(
    'div',
    attrs={
        'class': 'adProduct_info_area__dTSZf'
    }
)
```

```python
def find_data(x):
    # x는 beautifulsoup data
    result = x.find(
        'div',
        attrs={
            'class': 'adProduct_title__amInq'
        }
    )
    return result.get_text()

product_list = list(
    map(
        find_data,
        ad_list
    )
)
```

```python
ad_price_list = list(
    map(
        lambda x : x.find(
            'div',
            attrs={
                'class': 'adProduct_price_area__yA7Ad'
            }
        ),
        ad_list
    )
)
```

```python
ad_price_list
```

```python
# 비어있는 리스트 생성
price_list = []

for i in ad_price_list:
    result = i.find('em').get_text()
    price_list.append(result)
price_list
```

```python
data = div_data.find_all(
    'div',
    attrs={
        'class': 'product_title__Mmw2K'
    }
)
```

```python
for i in data:
    result = i.get_text()
    product_list.append(result)
product_list
```

```python
data2 = div_data.find_all(
    'div',
    attrs={
        'class': 'product_price_area__eTg7I'
    }
)
```

```python
data3 = list(
    map(
        lambda x : x.find(
            'em'
        ).get_text(),
        data2
    )
)
```
